```python
import matplotlib.pyplot as plt
import pandas as pd

lowbirth_df = pd.read_csv('./project2.csv', index_col=0)

plt.scatter(lowbirth_df['서울매매가격지수'], lowbirth_df['출생아수'], color = ['#54CAEA'])
plt.title('서울매매가격지수와 출생아수의 관계')
plt.xlabel('서울매매가격지수(2017.11=100기준)')
plt.ylabel('출생아수')
plt.show()
```
